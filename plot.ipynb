```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = []
names = ['monte', 'mult', 'inv']
for name in names:
    tmp = []
    for i in [1,2,4,8,16]:
        with open (name+'_'+str(i)+'.txt', 'r') as file:
            raw = file.read().split('\n')

        if name != 'inv': tmp.append(np.array([float(x.split()[1]) for x in raw[1:-1][1::2]]))
        else: tmp.append(np.array([float(x.split()[1]) for x in raw[1:-1][1::2]]))
    data.append(tmp)
```

```python
data = [np.array(x) for x in data]
```

```python
data
```

```python
fig, ax = plt.subplots(3,1, figsize=(12,15))
plt.subplots_adjust(hspace=1.)
plt.rcParams.update({'font.size': 18})
labels = str([1,2,4,8,16])
for i, name in enumerate(names):
    plt.subplot(3,1,i+1)
    plt.title(name)
    plt.xlabel('$2^N$')
    plt.ylabel('Time (s)')
    if name=='monte': plt.yscale('log')
    for j, key in enumerate([1,2,4,8,16]): plt.plot(data[i][j], label='P='+str(key))
    plt.legend()
```
